### Funcionamiento de Ver Datos del Perfil

1. °URL: **/accounts/profiles/** **GET**
    
    Devuelve el html para ver la información de todos los perfiles si es administrador, de lo contrario, redireccionará al **Home**.

2. °Variables que envía: **GET**

    - users_list: Esta lista devuelve la siguiente información de todos los usuarios:

        - identification: Tipo Int

        - first_name: Tipo Char(50)

        - last_name: Tipo Char(50)

        - email: Tipo Char(255)

        - phone_number: Tipo Char(50)

        - salary: Tipo Float _(Puede ser nulo así que se debe validar que rol tiene el usuario)_

        - entrance_date: Tipo Datetime

        - is_staff: Tipo Bool _(Si es empleado o no)_

        - is_superuser: Tipo Bool _(Si es admin o no)_

        - charge_name: Tipo Char _(Puede ser nulo así que se debe validar que rol tiene el usuario)_

    - errors: Es una lista de diccionarios con los errores que se hayan realizado, cada registro de la lista es un diccionario con las llaves:

        - title: Título del error.

        - content: Descripción del error.


3. °Variables que recibe:

4. Nombre y ruta del HTML: **user/profiles.html**


    

### Funcionamiento de Obtener datos de un usuario en específico

1. °URL: **/accounts/get_profile/** **POST**
    
    Devuelve el html para ver la información del perfil seleccionado por el admin. Si no es administrador redireccionará al **Home**.

2. °Variables que envía: **POST**

    - user_info: Diccionario con los siguientes datos del usuario:

        - identification: Tipo Int

        - first_name: Tipo Char(50)

        - last_name: Tipo Char(50)

        - email: Tipo Char(255)

        - phone_number: Tipo Char(50)

        - salary: Tipo Float _(Puede ser nulo así que se debe validar que rol tiene el usuario)_

        - is_staff: Tipo Bool _(Si es empleado o no)_

        - is_superuser: Tipo Bool _(Si es admin o no)_

        - charge_id: Tipo Int _(Puede ser nulo así que se debe validar que rol tiene el usuario)_

    - charge_list: Lista de diccionarios de cargos con las llaves: 'id', 'name' 

    - errors: Es una lista de diccionarios con los errores que se hayan realizado, cada registro de la lista es un diccionario con las llaves:

        - title: Título del error.

        - content: Descripción del error.


3. °Variables que recibe: (Para guardar datos) **POST**

    - identification: Tipo Int

4. Nombre y ruta del HTML: **user/profile.html**
    

### Funcionamiento de Guardar los datos de un usuario en específico.

1. °URL: **/accounts/save_profile/** **POST AJAX - NO FORM**
    
    Devuelve un JSON para validar si se guardaron los datos correctamente o no. Si no es administrador llega vacío todo.

2. °Variables que envía: **POST AJAX - NO FORM**

    - errors: Es una lista de diccionarios con los errores que se hayan realizado, cada registro de la lista es un diccionario con las llaves:

        - title: Título del error.

        - content: Descripción del error.
        
    - message: Diccionary _(Permite validar si la actualización del perfil fue correcta o no.)_

        El diccionario si no viene vacío, tiene lo siguiente:

        - title: Título del mensaje _(Actualización de perfil correcto o incorrecto)_

        - content: Contenido del mensaje _(Breve descripción de lo que sucedió)_

3. °Variables que recibe: (Para guardar datos) **POST AJAX - NO FORM**

    - identification: Tipo Int

    - first_name: Tipo Char(50)

    - last_name: Tipo Char(50)

    - email: Tipo Char(255)

    - phone_number: Tipo Char(50)

    - salary: Tipo Float or Null

    - is_staff: Tipo Bool _Si es empleado o no_

    - is_superuser: Tipo Bool _Si es admin o no_

    - charge_id: Tipo Int or Null
    
4. Nombre y ruta del HTML: **No tiene porque es JSON**
